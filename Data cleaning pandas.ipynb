```python
import pandas as pd
```

```python
df = pd.read_excel("Customer Call List.xlsx")
df
```

```python
#removing duplicates
#assign it back to df in order to save the changes
#you can also write inplace=true in order to save the changes 
df = df.drop_duplicates()
df
```

```python
#deleting columns that are non needed/useless
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
#stripping unnecessary characters from the strings/names of people
#strip() will take out characters from both sides of the string
#lstrip() will take out from the left
#rstrip() will take out form the right 
#you have to assign this back to  --- otherwise, it will make changes to every column. But you just want it on that one column
#you can specify within the brackets which character to take out, you can right multiple of them, without a comma
df["Last_Name"] = df["Last_Name"].str.strip("./")
df  #this will print out the dataframe with the changes made only to the lastname column
```

```python
#cleaning/standardising phone numbers
#firstly take out the forward slashes
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)  # Remove non-alphanumeric characters
df['Phone_Number']
```

```python
df.dtypes
```

```python
#we need to convert it to string before we can use lamda to format it. Currently the phone number column is an object

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))  # Convert to string
```

```python
#format to inlcude a dash after every three digits

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])# Format as XXX-XXX-XXXX
df["Phone_Number"]
```

```python
#taking out the nan--- and Na--

df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')

df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')

df
```

```python
#splitting the address column into three new ones
#were splitting the address by the comma
#were naming the three columns we've created

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand=True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')

df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')

df
```

```python
df=df.fillna('')
df
```

```python
#dropping a row
#were looping through the 'do not contact' column and if it says Y then we drop it 
#because the person doesn't want to be contacted. and we do this using the index
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
#dropping rows
#in the phone number column, were dropping all rows without
#a phone number in order to leave just people who provided their phone number and those who said that thye want to be called 

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df


#or method 2 another way to drop null values

#df = df.dropna(subset = ["Do_Not_Contact"]), inplace=True
```

```python
#dropping the index
#resetting it to chronological order

df = df.reset_index(drop=True)
df
```
